# Vader sentiment analysis

## Vader function

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
from string import digits
import pandas as pd 
import io 
import codecs
import csv
import re
import string
```

```python
analyser = SentimentIntensityAnalyzer()

def sentiment_analyzer_scores(sentence):
    score = analyser.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(score)))
    return score
```

## Import dataset

```python
Boeing = pd.read_csv('../social_data/reddit/boeing_reddit.csv')
Boeing = Boeing.dropna()
```

```python
stop_words = set(stopwords.words('english')) 
file1 = codecs.open('../social_data/reddit/boeing_reddit.csv','r','utf-8') 
line = file1.read() 
words = line.split()
for r in words: 
    if not r in stop_words: 
        appendFile = open('boeing_reddit.csv','a', encoding='utf-8') 
        appendFile.write(r)
        appendFile.write("\n")
        appendFile.close()
```

```python
def preprocessing(x):
    x = x.lower()
#     x = x.replace('—', ' ')
    x = x.replace('-', ' ')
    x = re.sub('\w+@\w+\.[a-z]{3}', ' ', x)
    x = x.translate(str.maketrans(" ", " ", string.punctuation))
    x = x.translate(str.maketrans('', '', digits))
    x = x.replace('“', '')
    x = x.replace('”', '')
    x = x.replace('…', '')
    x = x.replace('’', '')
    x = x.replace('‘', '')  
    x = ' '.join(x.split())
    x = x.split('—')
#     print(x)
    if(x[0].isupper()):
        x = x[1:]
    x = ' '.join(x)
    x = x.strip()
    x = x.replace('  ', '')
    return x
Boeing['comments'].apply(preprocessing)
```

```python
def get_stemmed_text(corpus):
    from nltk.stem.porter import PorterStemmer
    stemmer = PorterStemmer()
    return [' '.join([stemmer.stem(word) for word in review.split()]) for review in corpus]

Boeing['comments'].apply(get_stemmed_text)
Boeing.head()
```

```python
comments = Boeing['comments']
counter = 0
positive_sentiment_array = []
negative_sentiment_array = []
sentiment_score_array = []
label_array = []
topic_array = []
for comment in comments:
    counter += 1
#     obj = TextBlob( comment )
    sentiment = sentiment_analyzer_scores(comment)
#     sentiment = obj.sentiment.polarity
    if sentiment['compound'] > 0.0:
#             positive_sentiment_array.append(obj)
            label_array.append("positive")
    if sentiment['compound'] <0.0:
#             negative_sentiment_array.append(obj)
            label_array.append("negative")
    if sentiment['compound'] == 0.0:
            label_array.append('neutral')
    sentiment_score_array.append(sentiment['compound'])
    topic_array.append("Boeing Crisis")
    
# returns the sentiment of text
# by returning a value between -1.0 and 1.0

print(len(positive_sentiment_array))
print(len(negative_sentiment_array))
print(counter)
```

```python
Boeing.insert(1, "Score", sentiment_score_array , True) 
Boeing.insert(2, "Label", label_array , True) 
Boeing.insert(3, "Topic", topic_array , True) 
```

```python
Boeing = Boeing[Boeing['comments'] != '[deleted]']
```
